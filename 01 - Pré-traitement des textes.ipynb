```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import string
import re
```

```python
# ignorer les messages d'evertissement
import warnings
warnings.filterwarnings('ignore')
```

```python
# Charger les données
df = pd.read_csv('./anthems.csv')
```

```python
# Vérifier les données
print(df.head())
```

# 1. Nettoyage Superficiel

```python
# la colonne contenant les hymnes s'appelle 'Anthem'
anthems = df['Anthem']


def clean_text(text):
    # Convertir en minuscules
    text = text.lower()
    # Supprimer les balises HTML
    text = re.sub(r'<.*?>', '', text)
    # Supprimer les caractères spéciaux et les chiffres
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Supprimer les espaces supplémentaires
    text = text.strip()
    return text

anthems_cleaned = anthems.apply(clean_text)
```

# 2. Nettoyage Sémantique

```python
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def lemmatize_text(text):
    tokens = nltk.word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    return ' '.join(lemmatized_tokens)

anthems_lemmatized = anthems_cleaned.apply(lemmatize_text)
```

# 3. Tokenisation et Vectorisation

```python
# 3. Tokenisation
# Cette étape est intégrée dans la vectorisation avec TfidfVectorizer


vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(anthems_lemmatized)

# Vérifier la forme de la matrice de caractéristiques
print(X.shape)

# Sauvegarder la matrice vectorisée pour une utilisation future
import pickle
with open('./anthems_vectorized.pkl', 'wb') as f:
    pickle.dump(X, f)
```
